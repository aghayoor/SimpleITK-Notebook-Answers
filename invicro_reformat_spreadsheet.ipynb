```python
outDataFrame = DataFrame.loc[(DataFrame.SeriesDesc == 'AD')]
outDataFrame
```

```python
# rename columns
outDataFrame = outDataFrame.rename(columns={'lMLF Mean (UNK)':'AD lMLF Mean (mm^2/s)', 
                                            'lMLF_lesion Mean (UNK)':'AD lMLF_lesion Mean (mm^2/s)',
                                            'lMLF_nonlesion Mean (UNK)':'AD lMLF_nonlesion Mean (mm^2/s)',
                                            'rMLF Mean (UNK)':'AD rMLF Mean (mm^2/s)',
                                            'rMLF_lesion Mean (UNK)':'AD rMLF_lesion Mean (mm^2/s)',
                                            'rMLF_nonlesion Mean (UNK)':'AD rMLF_nonlesion Mean (mm^2/s)',})
```

```python
outDataFrame.columns
```

```python
## Add FA
FA_DF = DataFrame.loc[(DataFrame.SeriesDesc == 'FA')]
fa_lmlf = FA_DF[['lMLF Mean (UNK)']].values
fa_lmlf_lesion = FA_DF[['lMLF_lesion Mean (UNK)']].values
fa_lmlf_nonlesion = FA_DF[['lMLF_nonlesion Mean (UNK)']].values
fa_rmlf = FA_DF[['rMLF Mean (UNK)']].values
fa_rmlf_lesion = FA_DF[['rMLF_lesion Mean (UNK)']].values
fa_rmlf_nonlesion = FA_DF[['rMLF_nonlesion Mean (UNK)']].values

outDataFrame['FA lMLF Mean (UNK)'] = pd.Series(fa_lmlf.flatten(), index=outDataFrame.index)
outDataFrame['FA lMLF_lesion Mean (UNK)'] = pd.Series(fa_lmlf_lesion.flatten(), index=outDataFrame.index)
outDataFrame['FA lMLF_nonlesion Mean (UNK)'] = pd.Series(fa_lmlf_nonlesion.flatten(), index=outDataFrame.index)
outDataFrame['FA rMLF Mean (UNK)'] = pd.Series(fa_rmlf.flatten(), index=outDataFrame.index)
outDataFrame['FA rMLF_lesion Mean (UNK)'] = pd.Series(fa_rmlf_lesion.flatten(), index=outDataFrame.index)
outDataFrame['FA rMLF_nonlesion Mean (UNK)'] = pd.Series(fa_rmlf_nonlesion.flatten(), index=outDataFrame.index)
```
