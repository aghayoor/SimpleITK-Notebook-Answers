# What is the order of a CompositeTransform applies it Transform?
# What are the implications of the "Center" transform parameter for Optimization?

To answer these questions we will render a basic 2D image with an interacive slicer to apply a rotation transform under different conditions.

```python
import SimpleITK as sitk
print sitk.Version()

from myshow import myshow, myshow3d

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from myshow import myshow
from IPython.html import widgets
from IPython.html.widgets import interact, interactive
from math import pi
```

```python
img = sitk.ReadImage("Data/cthead1.png")
myshow(img, title="Input Image")
```

```python
def callback(transform, angle, img, title):
    """A call back method to resample an image with a active 2D Euler Transform"""
    # Here we use the old SimpleITK 0.8 transform inferface
    p = list(transform.GetParameters()) 
    p[0] = angle 
    transform.SetParameters(p)
    f = sitk.ResampleImageFilter()
    f.SetTransform(transform)
    f.SetReferenceImage(img)
    out = f.Execute(img)
    
    myshow(sitk.Compose(out, out, sitk.Maximum(out,img)), title=title)
```

Here we are using the old 0.8 Transform interface which only exposed the Get/Set Parameters and FixedParamters. New with 0.9 typed transforms are available with transform specific interface. So a "SetAngle" method could be used with 0.9.

```python
tx = sitk.Transform(2,sitk.sitkEuler)

    
interact(lambda **kwargs: callback(tx, img=img, title="Origin Center Of Rotation", **kwargs),             
             angle=widgets.FloatSliderWidget(min=-pi, max=pi, step=.02, value=0.0)
             )
```

The Above interaction is default state of a EulerTransform where the rotaion occours around the image's origin.

```python
tx2 = sitk.Transform(2,sitk.sitkEuler)
tx2.SetFixedParameters([45.8801, 48.03]) # SetCenter from CenterTransformInitializer
    
interact(lambda **kwargs: callback(tx2, img=img, title="Mass Center Of Rotation", **kwargs),
             angle=widgets.FloatSliderWidget(min=-pi, max=pi, step=.02, value=0.0) )
```

## Composite transform with optional Center

Below compose a composite transform with a translation of the center of mass to origin, followed by the Euler transform with the interactive widget again connected to the widget.

The Composite transform only exposes the Eulers Fixed and optimizable parameters, hiding the transforms.

Setting the Euler's Center or Fixed Parameter can be explored....



```python
ctx1 = sitk.Transform(2,sitk.sitkTranslation)
ctx1.SetParameters([45.8801, 48.03]) # SetTranslation

ctx2 = sitk.Transform(2,sitk.sitkEuler)
#ctx2.SetFixedParameters([45.8801, 48.03]) # SetCenter

ctx = sitk.Transform(ctx1)
ctx.AddTransform(ctx2)

interact(lambda **kwargs: callback(ctx, img=img, title="Composite Transform", **kwargs),
             angle=widgets.FloatSliderWidget(min=-pi, max=pi, step=.01, value=0.0) )
```
